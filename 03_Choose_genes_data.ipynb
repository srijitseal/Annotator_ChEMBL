```python
import pandas as pd
```

```python
final_df = pd.read_csv("Binding_activities_chembl_mitotox_cleaned.csv.gz", compression='gzip')
```

```python
final_df
```

```python
import numpy as np

final_df['label'] = np.where(final_df['activity_comment_class'] != 'inactive', 'active', 'inactive')
final_df
```

```python
# First sort by 'activity_comment_class' so that 'active' comes before 'inactive'
final_df = final_df.sort_values(by=['gene_symbol', 'parent_molecule_chembl_id', 'activity_comment_class'], ascending=[True, True, False])

# Then, drop duplicates based on 'gene_symbol' and 'parent_molecule_chembl_id'
final_df = final_df.drop_duplicates(subset=['gene_symbol', 'parent_molecule_chembl_id'], keep='first').reset_index(drop=True)
final_df
```

```python
counts = final_df.groupby('gene_symbol')['label'].value_counts().unstack().fillna(0)
# Print the counts
filtered_counts = counts[(counts['active'] >= 1) | (counts['inactive'] >= 1)]
filtered_counts.reset_index()
filtered_counts
```

```python
final_df = final_df[final_df['gene_symbol'].isin(filtered_counts.index.to_list())].reset_index(drop=True)
final_df
```

```python
for i in final_df.parent_molecule_chembl_id.unique():
    
    print(i)
```

```python
import matplotlib.pyplot as plt

# Sum the active and inactive compounds
totals = filtered_counts.sum()

# Plotting
totals.plot(kind='bar', color=['green', 'red'], figsize=(8, 6))

# Enhancing the plot for better readability
plt.ylabel('Count')
plt.title('Total Number of Actives (Green) and Inactives (Red) Across All Genes')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Calculate total compounds per gene
total_per_gene = filtered_counts.sum(axis=1).sort_values(ascending=False)
total_per_gene
```

```python
import matplotlib.pyplot as plt

# Plotting
plt.figure(figsize=(5, 5))
plt.hist(total_per_gene,log=False, color='grey', edgecolor='black')

# Enhancing the plot for better readability
plt.ylabel('Log Frequency')
plt.xlabel('Number of Compounds')
plt.title('Distribution of Total Number of Compounds (Actives + Inactives) Per Gene')

plt.tight_layout()
plt.show()
```

```python
final_df_matrix = final_df[["molecule_chembl_id", "gene_symbol", "label"]].drop_duplicates(subset=["molecule_chembl_id", "gene_symbol"])
final_df_matrix
```

```python
pivoted_df = final_df_matrix.pivot(index='molecule_chembl_id', columns='gene_symbol', values='label')
pivoted_df
```

```python
pivoted_df.replace({"inactive":0}, inplace=True)
pivoted_df.replace({"active":1}, inplace=True)
```

```python
pivoted_df_v2 = pivoted_df.reset_index().reset_index(drop=True)
pivoted_df_v2
```

```python
jumpcp = pd.read_csv("mitotox_chemblidentifiers.csv.gz", compression="gzip")
jumpcp = jumpcp[jumpcp.ChEMBL_ID.notna()].reset_index(drop=True).drop_duplicates(subset="Smiles")
jumpcp
```

```python
final_target_data = pd.merge(jumpcp, pivoted_df_v2,right_on="molecule_chembl_id", left_on="ChEMBL_ID")
final_target_data
```

```python
final_target_data.to_csv("targetannotations_singleproteins_mitotoxcompounds_25.csv", index=False)
```
